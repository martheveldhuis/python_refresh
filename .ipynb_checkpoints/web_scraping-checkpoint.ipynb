## Web scraping

```python
import requests
import bs4
```

```python
website = "http://quotes.toscrape.com/"
res = requests.get(website)
```

### Get the names of all authors on the page (https://www.crummy.com/software/BeautifulSoup/bs4/doc/) 

```python
soup = bs4.BeautifulSoup(res.text, "lxml")
print(soup.prettify())
```

```python
soup.select(".author")
```

```python
authors = set()

for name in soup.select(".author"):
    authors.add(name.text)
```

```python
authors
```

```python
quotes = []

for quote in soup.select(".text"):
    quotes.append(quote.text)
```
